In this assignment we will look at the bivariate normal distribution and try to find some of its properties and the effects of the correlation coefficient and the covariance matrix on it. First we will derive the bivariate normal from the regular normal distribution. $\\$
Let $Z_{1}$ and $Z_{2}$ be two independent normal distributions $N(0,1)$ and two normal random variables $X$ and $Y$ with means $\mu_{X},\space \mu_{Y}$, variances $\sigma_{X}^{2},\space \sigma_{Y}^{2}$ and correlation coefficient $\rho$ defiend as: $\\$
$$X=\sigma_{X}Z_{1}+\mu_{X}$$ 
$$Y=\sigma_{Y}(\rho Z_{1}+ \sqrt{1-\rho^{2}}Z_{2}) + \mu_{Y}$$


a. Show that $X$ and $Y$ are bivariate normal and find the joint PDF $f(x,y)$ by using the method of transformation for continuous random variables.
   (show that $aX+bY$ has a normal distribution $\forall  a,b \in \mathbb{R} $) $\\$

Ans: $\\$

$aX + bY = a(\sigma_{X}Z_{1}+\mu_{X}) + b(\sigma_{Y}(\rho Z_{1}+ \sqrt{1-\rho^{2}}Z_{2}) + \mu_{Y}) = (a\sigma_{X}+ b\sigma_{Y}\rho)Z_{1} + b\sigma_{Y}\sqrt{1-\rho^{2}}Z_{2} + a\mu_{X} + b\mu_{Y}$ $\\$ 
The sum of two normal distributions is a normal distribution with different mean and variance as a result of multiplication with a constant, the addition of the constant shifts the mean by that value. $\\$

To find the joint PDF by method of transformations, we first need to defind $Z_{1}$ and $Z_{2}$ as functions of $X,Y$: 

$$Z_{1} = \frac{X-\mu_{X}}{\sigma_{X}} = g_{1}(X,Y)$$
$$Z_{2} = \frac{Y-\mu_{Y}}{\sigma_{Y}\sqrt{1-\rho^{2}}} - \frac{\rho Z_{1}}{\sqrt{1-\rho^{2}}} = \frac{Y-\mu_{Y}}{\sigma_{Y}\sqrt{1-\rho^{2}}} - \frac{\rho (X - \mu_{X})}{\sigma_{X}\sqrt{1-\rho^{2}}} = g_{2}(X,Y)$$

$Z_{1}$ and $Z_{2}$ are independent, so they have joint PDF as followed: $%:$

$$f_{Z_{1}Z_{2}}(z_{1}, z_{2}) = f_{Z_{1}}(z_{1})f_{Z_{2}}(z_{2}) = \frac{1}{2\pi}e^{-\frac{1}{2}z_{1}^{2}}e^{-\frac{1}{2}z_{2}^{2}} =\frac{1}{2\pi}e^{-\frac{1}{2}(z_{1}^{2} + z_{2}^{2})}$$

from method of transformation we know that $%:$

$$f_{XY}(x,y)=f_{Z_{1}Z_{2}}(g_{1}(x,y), g_{2}(x,y))|\bold{J}|$$

where $|\bold{J}|$ is the determinant of the Jacobian matrix $%:$

$$ \bold{J} = \left(\begin{array}{cc} \frac{\partial g_1}{\partial x} & \frac{\partial g_1}{\partial y} \\ \frac{\partial g_2}{\partial x} & \frac{\partial g_2}{\partial y} \end{array} \right) \rightarrow |\bold{J}| = \frac{\partial g_1}{\partial x} \frac{\partial g_2}{\partial y} - \frac{\partial g_1}{\partial y} \frac{\partial g_2}{\partial x}$$
$$\bold{J} = \left(\begin{array}{cc} \frac{1}{\sigma_{X}} & 0 \\ -\frac{\rho}{\sigma_{X}\sqrt{1-\rho^{2}}} & \frac{1}{\sigma_{Y}\sqrt{1-\rho^{2}}} \end{array} \right) \rightarrow |\bold{J}| = \frac{1}{\sigma_{X}\sigma_{Y}\sqrt{1-\rho^{2}}}$$

thus we get $%:$

$$\begin{align*}
f_{XY}(x,y) &= f_{Z_{1}Z_{2}}\left(\frac{x-\mu_{X}}{\sigma_{X}},\frac{y-\mu_{Y}}{\sigma_{Y}\sqrt{1-\rho^{2}}} - \frac{\rho (x - \mu_{X})}{\sigma_{X}\sqrt{1-\rho^{2}}}\right) \centerdot \frac{1}{\sigma_{X}\sigma_{Y}\sqrt{1-\rho^{2}}}\\

&= \frac{1}{2\pi\sigma_{X}\sigma_{Y}\sqrt{1-\rho^{2}}}\exp\left(-\frac{1}{2}\left[\left(\frac{x-\mu_{X}}{\sigma_{X}}\right)^{2} + \left(\frac{y-\mu_{Y}}{\sigma_{Y}\sqrt{1-\rho^{2}}} - \frac{\rho (x - \mu_{X})}{\sigma_{X}\sqrt{1-\rho^{2}}} \right)^{2} \right] \right) \\

&= \frac{1}{2\pi\sigma_{X}\sigma_{Y}\sqrt{1-\rho^{2}}}\exp\left(-\frac{1}{2(1-\rho^{2})}\left[\left(\frac{x-\mu_{X}}{\sigma_{X}}\right)^{2}(1-\rho^{2}) + \left(\frac{y-\mu_{Y}}{\sigma_{Y}} - \frac{\rho (x - \mu_{X})}{\sigma_{X}} \right)^{2} \right] \right) \\

&= \frac{1}{2\pi\sigma_{X}\sigma_{Y}\sqrt{1-\rho^{2}}}\exp\left(-\frac{1}{2(1-\rho^{2})}\left[\left(\frac{x-\mu_{X}}{\sigma_{X}}\right)^{2}-2\rho\left(\frac{(x-\mu_{X})(y-\mu_{Y})}{\sigma_{X}\sigma_{Y}}\right) + \left(\frac{y-\mu_{Y}}{\sigma_{Y}}\right)^{2}\right]\right)
\end{align*}$$

Calculate the covariance matrix and show by using $$f(\bold{x})=\frac{1}{\sqrt{(2\pi)^{k}|det(\bold{\Sigma})|}}\exp\left({-\frac{1}{2}(\bold{x}-\bold{\mu})^{T}}\bold{\Sigma}^{-1}(\bold{x}-\bold{\mu})\right)$$ that we get the same joint distribution from section a, where $\bold{x}$ is a column vector of $x,y$ parameters of the random variables, $\bold{\mu}$ is a column vectors of the means $\mu_{X}, \mu_{Y}$,  $\bold{\Sigma}$ is the covariance matrix and $k$ is the dimension of $\bold{x}$. $\\$

Ans: $\\$

first we find the covariance matrix: $\\$

$$ \bold{\Sigma} = \left(\begin{array}{cc} \Cov(X,X) & \Cov(X,Y) \\ \Cov(X,Y) & \Cov(Y,Y) \end{array} \right) $$

We know that: $\Cov(X,X) = \left<(x-\mu_{X})(x-\mu_{X})\right> = \left<(x-\mu_{X})^{2}\right> = \Var(X) = \sigma_{X}^{2}$, similarlly $\Cov(Y,Y) = \sigma_{Y}^{2}$. We can find $\Cov(X,Y)$ by using the definition of pearson correlation coefficient 
$$\rho = \frac{\left<(x-\mu_{X})(y-\mu_{Y})\right>}{\sigma_{X}\sigma_{Y}} = \frac{\Cov(X,Y)}{\sigma_{X}\sigma_{Y}} \rightarrow \Cov(X,Y) = \rho\sigma_{X}\sigma_{Y} $$

The covariance matrix is:
$$ \bold{\Sigma} = \left(\begin{array}{cc} \sigma_{X}^{2} & \rho\sigma_{X}\sigma_{Y} \\ \rho\sigma_{X}\sigma_{Y} & \sigma_{Y}^{2} \end{array} \right) \rightarrow det(\Sigma) = \sigma_{X}^{2}\sigma_{Y}^{2}(1-\rho^{2}) $$

The term outside the exponent for $k=2$ will be:
$$\begin{align*}
\frac{1}{\sqrt{(2\pi)^{k}|det(\bold{\Sigma})|}} = \frac{1}{\sqrt{(2\pi)^{2}\sigma_{X}^{2}\sigma_{Y}^{2}(1-\rho^{2})}} = \frac{1}{2\pi\sigma_{X}\sigma_{Y}\sqrt{1-\rho^{2}}}
\end{align*}$$
which is similar to the term outside the exponent of the joint PDF we found in section a.

In case of a 2x2 matrix, the inverse matrix can be calculated as follow:
$$ A = \left(\begin{array}{cc} a & b \\ c & d \end{array} \right) \rightarrow A^{-1} = \frac{1}{det(A)}\left(\begin{array}{cc} d & -b \\ -c & a \end{array} \right) \quad;\quad \mathrm{for} \space det(A) \neq 0 $$

for the inverse covariance matrix we get:
$$ \bold{\Sigma}^{-1} = \frac{1}{\sigma_{X}^{2}\sigma_{Y}^{2}(1-\rho^{2})}\left(\begin{array}{cc} \sigma_{Y}^{2} & -\rho\sigma_{X}\sigma_{Y} \\ -\rho\sigma_{X}\sigma_{Y} & \sigma_{X}^{2} \end{array} \right) $$

now we can calculate the expression in the exponent for $\bold{\mu} = \left( \begin{array}{c} \mu_{X} \\ \mu_{Y} \end{array} \right)$ and $\bold{X} = \left( \begin{array}{c} x \\ y \end{array} \right)$:

$$\begin{align*}
-\frac{1}{2} (\bold{x}-\bold{\mu})^{T}\bold{\Sigma}^{-1}(\bold{x}-\bold{\mu}) &= \left( \begin{array}{c} x-\mu_{X} ,& y-\mu_{Y} \end{array} \right) \frac{1}{\sigma_{X}^{2}\sigma_{Y}^{2}(1-\rho^{2})} \left(\begin{array}{cc} \sigma_{Y}^{2} & -\rho\sigma_{X}\sigma_{Y} \\ -\rho\sigma_{X}\sigma_{Y} & \sigma_{X}^{2} \end{array} \right) \left( \begin{array}{c} x-\mu_{X} \\ y-\mu_{Y} \end{array} \right) \\

&= - \frac{1}{2\sigma_{X}^{2}\sigma_{Y}^{2}(1-\rho^{2})} \left( \begin{array}{c} (x-\mu_{X})\sigma_{Y}^{2} - (y-\mu_{Y})\rho\sigma_{X}\sigma_{Y} ,& (y-\mu_{Y})\sigma_{X}^{2} - (x-\mu_{X})\rho\sigma_{X}\sigma_{Y} \end{array} \right) \left( \begin{array}{c} x-\mu_{X} \\ y-\mu_{Y} \end{array} \right) \\

&= - \frac{1}{2\sigma_{X}^{2}\sigma_{Y}^{2}(1-\rho^{2})} \left((x-\mu_{X})^{2}\sigma_{Y}^{2} - (y-\mu_{Y})(x-\mu_{X})\rho\sigma_{X}\sigma_{Y}  + (y-\mu_{Y})^{2}\sigma_{X}^{2} - (y-\mu_{Y})(x-\mu_{X})\rho\sigma_{X}\sigma_{Y}\right) \\

&= - \frac{1}{2(1-\rho^{2})} \left(\left(\frac{x-\mu_{X}}{\sigma_{X}}\right)^{2} -2\rho\frac{(x-\mu_{X})(y-\mu_{Y})}{\sigma_{X}\sigma_{Y}} + \left(\frac{y-\mu_{Y}}{\sigma_{Y}}\right)^{2}\right)
\end{align*}$$

and we got the expression inside the exponent for the joint PDF in section a. This conclude that we got the same joint PDF. $\\$

d. To see how the correlation coefficient affects the bivariate normal, simulate a bivariate normal distribution and plot the distribution for different values $\rho = {0, 0.3, 0.5, 0.7, 0.9, 1, -0.4, -0.6, -0.8, -1}$. Assume $\sigma_{x}=\sigma_{y}=1$ and $\mu_{X}=\mu_{Y}=0$. Use the covariance matrix. $\\$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.stats import multivariate_normal
```

```python
rho_values = np.array([[0, 0.3, 0.5, 0.7, 0.9], 
                       [1, -0.4, -0.6, -0.8, -1]])
sig_x = 1 ; mean_x = 0
sig_y = 1 ; mean_y = 0
N = 2000
means_vector = np.array([mean_x, mean_y])

fig, axs = plt.subplots(2,5, figsize=(20,10)) 

for n in range(2): # n,m from the positions of the subplots axis
    for m in range(5):
        rho = rho_values[n,m] 
        cov_mat = np.array([[sig_x**2, rho*sig_x*sig_y],
                            [rho*sig_x*sig_y, sig_y**2]]) # calculate the covariance matrix
        pdf = multivariate_normal(means_vector, cov_mat, allow_singular=True) # define the bivariate normal
        data = pdf.rvs(N) # draws N sample from the distribution
        axs[n,m].scatter(data[:,0],data[:,1], marker='.')
        axs[n,m].set_title(r'$\rho$='+f'{rho}')
```

f. Choose one of the covariance matrices from previous section and preform PCA (Principal Components Analysis) whitening. $\\$

Ans: $\\$

For this section ill choose $\bold{\Sigma}_{1} = \left(\begin{array}{cc} 9 & 8 \\ 8 & 9 \end{array} \right)$, but this will work for the other matrices as well. $\\$
Our goal is to get an identity covariance matrix ($Var(X)=Var(Y)=1$) to whitening the data to get multivariate normal distribution white noise with mean 0 and variance 1. $\\$
First we calculate the covariance matrix's eigenvalues ($\lambda_{1}$ and $\lambda_{2}$) and eigenvectors ($\bold{v_{1}}$ and $\bold{v_{2}}$) to diagonalized the covariance matrix by applying:  $P^{-1}\Sigma P = D$, where P is a matrix of the eigenvectors as columns and D is diagonal matrix with the eigenvalues. By doing this step we decorrelate the data (rotating the covariance matrix by P). To scale the covariance matrix to get the identity covariance matrix we can use the relation of $D^{-1}D=I$ and $D^{-1}=D^{-1/2}ID^{-1/2}$ to get $D^{-1/2} P^{-1}\Sigma P D^{-1/2} = I$ we can define the matrix $W=D^{-1/2}P^{-1}$ as the whitening transformation matrix $W\Sigma W^{-1} = I$ to get the identity covariance matrix.

```python
# Define the covariance matrix and the means vector:
# sig1 = np.array([[9, 8],
#                  [8, 9]]) # rho = 8/9 = 0.888
chosen_cov_mat = sig1
string_chosen_cov = r'$\Sigma_{1}$'
means = np.array([0,0]) # mu_X = mu_Y =0
print(f'Chosen covariance matrix: \n' + string_chosen_cov + f'=\n {chosen_cov_mat}')

# Define the bivariate normal and meshgrid:
pdfc = multivariate_normal(means, chosen_cov_mat)
x_values, y_values = np.meshgrid(np.linspace(-10, 10, 1000), np.linspace(-10, 10, 1000)) # might need to change the domains for diffferent matrices to see all of the distribution in the first plot.
pos = np.dstack((x_values, y_values))

# Diagonalize the covariance matrix
eigenvalues, eigenvectors = np.linalg.eigh(chosen_cov_mat)
print(f'eigenvalues: {eigenvalues},\neigenvectors:\n {eigenvectors}')

inverse_eigenvectors =  np.linalg.inv(eigenvectors)
diagonalized_ccm = (inverse_eigenvectors @ chosen_cov_mat @ eigenvectors).round(5)

# Create a multivariate normal distribution with diagonal covariance matrix
pdfc_diag = multivariate_normal(means, diagonalized_ccm)


# Scaling the covariance matrix by D^{-1/2} to get the identity covariance matrix (whitening)
root_diag_eigenvectors =  np.diag(1/eigenvalues**0.5)
identity_covariance_matrix = root_diag_eigenvectors.T @ diagonalized_ccm @ root_diag_eigenvectors
print(f'Identity covariance matrix:\n{identity_covariance_matrix}')

# Create a multivariate normal distribution with the whiten covariance matrix
pdfc_identity = multivariate_normal(means, identity_covariance_matrix)



figc, axsc = plt.subplots(3,1, figsize=(8,20))

axsc[0].contourf(x_values, y_values, pdfc.pdf(pos), cmap='viridis')
axsc[0].set_title('PDF from ' + string_chosen_cov)

# Plot the decorrelated data (rotating the covariance matrix by P)
axsc[1].contourf(x_values, y_values, pdfc_diag.pdf(pos), cmap='viridis')
axsc[1].set_title('decorrelate ' + string_chosen_cov)

# Plot the whiten data
axsc[2].contourf(x_values, y_values, pdfc_identity.pdf(pos), cmap='viridis')
axsc[2].set_title('whiten ' + string_chosen_cov)
```
